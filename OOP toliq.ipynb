```python
#class yordamida qiymatlar bilan ishlash


class Employee:
    
    def __init__(self, first, last, pay, time):
        self.first = first
        self.last = last
        self.pay = pay
        self.time = time
        self.email = first + '.' + last + '@company.com'

    def fullname(self):
        return '{} {} {}'.format(self.first, self.last, self.time)

emp_1 = Employee('Corey', 'Schafer', 50000, '14-02-2022')
emp_2 = Employee('Test', 'User', 60000, '2002')

print(emp_1)
print(emp_2)

emp_1.fullname()
emp_2.fullname()
```

```python
class Employee:
    
    raise_amount = 1.04
    
    def __init__(self, first, last, pay):
        self.first = first
        self.last = last
        self.pay = pay
        self.email = first + '.' + last + '@company.com'

    def fullname(self):
        return '{} {}'.format(self.first, self.last)
    
    def apply_raise(self):
        self.pay = int(self.pay * self.raise_amount)
        

emp_1 = Employee('Corey', 'Schafer', 50000)
emp_2 = Employee('Test', 'User', 60000)

print(emp_1.pay)
emp_1.apply_raise()

print(emp_1.pay)


```

## 2. OOP Inheritance  - Me'ros

```python
# OTA-ONA (Parent) yoki ASOS classni yaratamiz

class Inson:
    def __init__(self, ism, familiya):
        self.ism = ism
        self.familiya = familiya
        
    def printism(self):
        print(self.ism, self.familiya)
        
#Person class yordamida obyekt yaratish va keyin printism methodi yordamida chop etish:

x = Inson("Yoqub", "Nosirov")
x.printism()
```

```python
#BOLA (Child) yoki HOSILA classni yaratish

class Talaba(Inson): #bunda Student classi Inson classining barcha belgi va methodlarini me'ros qilib oladi 
    pass                     #pass function yordamida Talaba classida hech narsa e'lon qilmaganimizni bildirdik

y = Talaba("Malika", 'Azizova')
y.printism()
```

```python
#Yuqoridagi misolda Pass funksiyasi yordamida Talaba me'ros classida hech narsa e'lon qilmadik

# __init__ yordamida me'ros classni hosil qilish

class Talaba(Inson):
    def __init__(self, ism, familiya):
        pass
        #belgilar qo'shish
        
# child classga __init__ni qo'shadigan bo'lsak bu class Asos classni me'ros qilib olmaydi
# Child classdagi __init__ me'ros qilib olingan belgilarni qayta yozadi
```

```python
#Asos classni belgilarini me'ros qilib olish uchun esa 

class Inson:
    def __init__(self, ism, familiya):
        self.ism = ism
        self.familiya = familiya

    def printism(self):
        print(self.ism, self.familiya)

class Talaba(Inson):
    def __init__(self, ism, familiya):
        Inson.__init__(self, ism, familiya)          #_init_yordamida asos classning belgi va methodlari olindi

x = Talaba("O'tkir", "Zokirov")
x.printism()
```

#### Super() funksiyasi yordamida me'ros olish

```python
class Inson:
    def __init__(self, ism, familiya):
        self.ism = ism
        self.familiya = familiya

    def printism(self):
        print(self.ism, self.familiya)

class Talaba(Inson):
    def __init__(self, ism, familiya):
        super().__init__(ism, familiya)          #super() yordamida asos classning belgi va methodlari olindi, self shart emas

x = Talaba("O'tkir", "Zokirov")
x.printism()
```

```python
class Inson:
    def __init__(self, ism, familiya):
        self.ism = ism
        self.familiya = familiya

    def printism(self):
        print(self.ism, self.familiya)

class Talaba(Inson):
    def __init__(self, ism, familiya, yosh):  #Talaba classiga yosh deb nomlangan yangi belgi qo'shildi, ism -familiya esa me'ros
        super().__init__(ism, familiya)
        self.yosh = yosh
   
y = Talaba("O'tkir", "Zokirov", 18)
print(y.yosh)
```

## 3. Enkapsulatsiya 

```python
#misol uchun quyidagi misolda biz Hodimlar degan class yaratdik va unga ism va oylik o'zgaruchilarini e'lon qildik
#Xuddi shu class ichida ish() va ko'rsat() singari methodlarni yaratdik 

class Xodimlar:
    # konstruktor
    def __init__(self, ism, oylik, proyekt):
        # o'zgaruvchilar yoki data members
        self.ism = ism
        self.oylik = oylik
        self.proyekt = proyekt

    # method
    # Xodim ma'lumotlarini ko'rsatuvchi method
    def korsat(self):
        # umumiy o'zgaruvchiga kirish
        print("Ismi:", self.ism, 'Oyligi:', self.oylik)

    # method
    def ish(self):
        print(self.ism, 'ning proyekti', self.proyekt)

# Classning obyektini yaratish
xodim1 = Xodimlar('Oybek', 8000, 'Python CV')

# classning methodlarini chaqirish
xodim1.korsat()
xodim1.ish()
```

#### Public, Private va Protected qiymatlar

### Public qiymatlar 

```python
#Public o'zgaruvchilarga class ichidan va class tasharisidan murojaat qilish mumkin. Class e'lon qilingan paytda barcha 
#o'zgaruvchilar public sifatida e'lon qilinadi. 

class Xodim:
    # Konstructor
    def __init__(self, ism, oylik):
        # public qiymatlar
        self.ism = ism
        self.oylik = oylik

    # public methodlar
    def printxodim(self):
        # public qiymatlarga murojaat
        print("Ismi: ", self.ism, 'Oylik maoshi:', self.oylik)

# Class obyektlarini yaratish
xodim1 = Xodim('Azizbek', 10000)

# public qiymatlarga murojaat qilish
print("Ismi: ", xodim1.ism, 'Oylik maoshi:', xodim1.oylik)

#classning public methodlarini chaqirish
xodim1.printxodim()
```

### Private qiymatlar 

```python
# Private (shaxsiy) qiymatlar yoki o'zgaruvchilar faqatgina shu class ichida murojaat qilinishi mumkin
#o'zgaruvchini private deb e'lon qilishimiz uchun __ belgisi talab etiladi . 

class Xodim:
    # Konstructor
    def __init__(self, ismi, oylik):
        # public data member
        self.ismi = ismi
        # private member
        self.__oylik = oylik

#Class obyektini yaratish
xod1 = Xodim('Azizbek', 10000)

# private data memberlariga murojaat qilish
print('Salary:', xod1.__oylik)    #__oylik private qiymat shu sababdan ham unga class tashqarisidan murojaat qila olmaymiz.
```

```python
#bunday holatda public method yordamida private qiymatlarni chop etishimiz mumkin
class Xodim:
    # Konstructor
    def __init__(self, ismi, oylik):
        # public data member
        self.ismi = ismi
        # private member
        self.__oylik = oylik
    
    def printxodim(self):
        #private qiymatlarga class ichidan turib murojaat qilish mumkin
        print("Ismi: ", self.ismi, 'Oylik maoshi: ', self.__oylik)

#Class obyektini yaratish
xod1 = Xodim('Azizbek', 10000)

# public sifatida qiymatlarni chop etish
xod1.printxodim()
```

### Protected qiymatlar 

```python
#Protected qiymatlarga class ichidan va subclasslardan murojaat qilish mumkin. _ yordamida e'lon qilinadi

#Asos class

class Kompaniya():
    def __init__(self):
        #Protected qiymat
        self._proyekt = "Python"
            
#Child - Bola class

class Xodim(Kompaniya):
    def __init__(self, ism):
        self.ism = ism
        Kompaniya.__init__(self)
        
    def show(self):
        print("Xodim ismi: ", self.ism)
        print("Hozirda ma'sul proyekt nomi:", self._proyekt)
        
a = Xodim("Sherzod")
a.show()

print('Proyekt:', a._proyekt)
```

## Python dasturlash tilida getterlar va setterlar

```python
class Talaba:
    def __init__(self, ism, yosh):
        # private member
        self.ism = ism
        self.__yosh = yosh

    # getter method
    def get_yosh(self):
        return self.__yosh

    # setter method
    def set_yosh(self, yosh):
        self.__yosh = yosh

tal = Talaba('Denis', 16)

# Getter yordamida yosh qiymatini chop etish
print('Ismi:', tal.ism, tal.get_yosh())

# Setter yordamida yosh qiymatini o'zgartirish 
tal.set_yosh(31)

# Getter yordamida yosh qiymatini chop etish.
print('Ismi:', tal.ism, tal.get_yosh())
```

```python
# Boshqa misol: Enkapsulation yordamida ma'lumotni yashirish va qiymatlarni o'zgartirish uchun shartlar e'lon qilish

class Talaba():
    def __init__(self, ism, ball, yosh):
        
        #private member
        self.ism = ism
        
        #murojaat cheklangan private members
        #to'g'ridan to'g'ri ma'lumotlarni o'zgartirishni oldini olish uchun
        self.__ball = ball
        self.__yosh = yosh
        
    def show(self):
        print('Talaba ma\'lumotlari: ', self.ism, self.__ball)
        
    #getter methodi
        
    def get_ball(self):
        return self.__ball
    
    #ma'lumotni o'zgartirish uchun setter methodi
    #ma'lumotni o'zgaritish uchun shart
    
    def set_ball(self, qiymat):
        if qiymat > 100:
            print("Mavjud bo'lmagan ball kiritildi, iltimos to'g'ri ball kiriting")
        else:
            self.__ball = qiymat

hilola = Talaba ('Hilola', 80, 19)

#Qiymatni o'zgartirishdan oldin 
hilola.show()

#Shart yordamida qiymatni o'zgartirishga o'rinish
hilola.set_ball(120)
hilola.set_ball(130)
hilola.set_ball(1000)

hilola.set_ball(99)
hilola.show()
       
```
